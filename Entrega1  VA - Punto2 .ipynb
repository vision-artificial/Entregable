```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("res/Montaje.jpg", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img1 =cv2.imread("res/Luz no controlada.jpg", cv2.IMREAD_COLOR)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 =cv2.imread("res/Luz controlada.jpg", cv2.IMREAD_COLOR)
img2 =cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(30, 7))
fig.suptitle('Montaje, objeto con luz no controlada y luz controlada', fontsize=20)
ax1.set_title('Montaje')
ax1.imshow(img, aspect='auto')
ax2.set_title('Luz no controlada')
ax2.imshow(img1, aspect='auto')
ax3.set_title('Luz controlada')
ax3.imshow(img2, aspect='auto')
```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("res/Luz controlada.jpg", cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_R = img[:,:,0]
img_G = img[:,:,1]
img_B = img[:,:,2]
#Se crea el subplot 1x3
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(30, 7))
#Se separan en los canales R, G  y B. Se ponen títulos y se muestran las imágenes
fig.suptitle('Separacion de canales RGB', fontsize=20)
ax1.set_title('Canal R')
ax1.imshow(img_R, cmap='Reds', aspect='auto')
ax2.set_title('Canal G')
ax2.imshow(img_G, cmap='Greens', aspect='auto')
ax3.set_title('Canal B')
ax3.imshow(img_B, cmap='Blues', aspect='auto')

#Se genera un arreglo para imprimir las imágenes, 3x2, 3 imágenes y 3 histogramas asociados.
fig, arreglo_plots = plt.subplots(3,2, figsize=(15, 15))
#Título del cuadro en la posición (0,0)
arreglo_plots[0,0].set_title('Imagen canal R')
#Imprimir la imagen 1 en la posición (0.0) del arreglo
arreglo_plots[0,0].imshow(img_R)
#Título del cuadro en la posición (0,1)
arreglo_plots[0,1].set_title('histograma')
#Imprimir el histograma en la posición (0,1)
_ = arreglo_plots[0,1].hist(img[:,:,0].ravel(), bins=256, range=(0.0, 255.0))

#Análogamente para los demás casos, como se observa varía la posición asociada en el subplot
arreglo_plots[1,0].set_title('Imagen canal G')
arreglo_plots[1,0].imshow(img_G)
arreglo_plots[1,1].set_title('histograma')
_ = arreglo_plots[1,1].hist(img1[:,:,1].ravel(), bins=256, range=(0.0, 255.0))

arreglo_plots[2,0].set_title('Imagen canal B')
arreglo_plots[2,0].imshow(img_B)
arreglo_plots[2,1].set_title('Histograma')
_ = arreglo_plots[2,1].hist(img2[:,:,2].ravel(), bins=256, range=(0.0, 255.0))
```
